# 5 - Daily Return Calculation / Calcular retornos diários

```python
# Calcula a variação percentual do mercado em um período de X dias

report_return = market.pct_change(periods = 5)

report_return.head(10)
```

AES Brasil, Alupar, Banco do Brasil, Itaúsa, Klabin, SANEPAR, TAEE, Kilima Suno 30

```python

# Extração dos retornos percentuais mais recentes (último valor) para cada ativo específico

return_AES_Brasil = report_return["AES Brasil"].iloc[-1]
return_Alupar = report_return["Alupar"].iloc[-1]
return_Banco_do_Brasil = report_return["Banco do Brasil"].iloc[-1]
return_Itaúsa = report_return["Itaúsa"].iloc[-1]
return_Klabin = report_return["Klabin"].iloc[-1]
return_SANEPAR = report_return["SANEPAR"].iloc[-1]
return_TAEE = report_return["TAEE"].iloc[-1]
return_Kilima_Suno_30 = report_return["Kilima Suno 30"].iloc[-1]
```

```python

# Conversão dos retornos percentuais em strings formatadas para exibição
# Multiplica cada retorno percentual por 100, arredonda para 2 casas decimais e adiciona o símbolo de porcentagem

return_AES_Brasil = str(round(return_AES_Brasil * 100, 2)) + "%"
return_Alupar = str(round(return_Alupar * 100, 2)) + "%"
return_Banco_do_Brasil = str(round(return_Banco_do_Brasil * 100, 2)) + "%"
return_Itaúsa = str(round(return_Itaúsa * 100, 2)) + "%"
return_Klabin = str(round(return_Klabin * 100, 2)) + "%"
return_SANEPAR = str(round(return_SANEPAR * 100, 2)) + "%"
return_TAEE = str(round(return_TAEE * 100, 2)) + "%"
return_Kilima_Suno_30 = str(round(return_Kilima_Suno_30 * 100, 2)) + "%"
```

```python
# Função para colorir o texto com base no valor

def color_text(value):
    
    if value.startswith('-'):
        # Vermelho para valores negativos
        return f"\033[91m{value}\033[0m"
    
    else:
        # Verde para valores positivos
        return f"\033[92m{value}\033[0m"
```

```python
# Print das informações com a função de coloração
print("AES Brasil: " + color_text(return_AES_Brasil))
print("Alupar: " + color_text(return_Alupar))
print("Banco do Brasil: " + color_text(return_Banco_do_Brasil))
print("Itaúsa: " + color_text(return_Itaúsa))
print("Klabin: " + color_text(return_Klabin))
print("SANEPAR: " + color_text(return_SANEPAR))
print("TAEE: " + color_text(return_TAEE))
print("Kilima Suno 30: " + color_text(return_Kilima_Suno_30))
```
